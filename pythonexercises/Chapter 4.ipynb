**Exercise 1**

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

Enter Hours: 45
Enter Rate: 10
Pay: 475.0

```python
 # Define a function called compute_pay that takes two parameters: hours and rate
def compute_pay(hours, rate):
    # Calculate the gross pay
    if hours <= 40:  #The gross pay for the first 40 hours worked is calculated as 40 times the hourly rate.
        # If the number of hours worked is less than or equal to 40, pay the regular hourly rate
        pay = hours * rate
    else:
        # If the number of hours worked is greater than 40, pay 1.5 times the hourly rate for the hours worked above 40
        pay = 40 * rate + (hours - 40) * rate * 1.5 #The gross pay for the hours worked above 40 is calculated as the number of hours worked above 40 multiplied by 1.5 times the hourly rate.

    # Return the gross pay
    return pay

# Get the number of hours worked from the user
hours = float(input('Enter Hours: '))

# Get the hourly rate from the user
rate = float(input('Enter Rate: '))

# Call the compute_pay function with the hours and rate as arguments
pay = compute_pay(hours, rate)

# Print the gross pay
print('Pay:', pay)
```

**CHALLENGE**

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0

Run the program repeatedly to test the various different values for input.

```python
# Define a function called computegrade that takes a score as its parameter
def computegrade(score):
    # Determine the grade based on the score
    if score >= 97 and score <= 100: #If the score is between 97 and 100
        grade = '1.0' # the display is 1.0 grade
    elif score >= 93 and score < 97: #If the score is between 93 and 97
        grade = '1.25' # the display is 1.25 grade
    elif score >= 89 and score < 93: #If the score is between 89 and 93
        grade = '1.5' # the display is 1.5 grade
    elif score >= 85 and score < 89: #If the score is between 85 and 89
        grade = '1.75'  # the display is 1.75 grade
    elif score >= 81 and score < 85:  #If the score is between 81 and 85
        grade = '2.0' # the display is 2.0 grade
    elif score >= 77 and score < 81: #If the score is between 77 and 81
        grade = '2.25' # the display is 2.25 grade
    elif score >= 74 and score < 77: #If the score is between 74 and 77
        grade = '2.5' # the display is 2.5 grade
    elif score >= 0 and score < 74: #If the score is between 0 and 77
        grade = '3.0' # the display is 3.0 grade
    else:
        # If the score is not between 0 and 100 (inclusive), print an Invalid score
        print('Invalid score')
        grade = None

    # Return the grade as a string
    return grade

# Get the score from the user
score = float(input('Enter a score between 0 and 100: '))

# Call the computegrade function with the score as an argument
grade = computegrade(score)

# Print the grade
if grade is not None:
    print('Grade:', grade)
```
