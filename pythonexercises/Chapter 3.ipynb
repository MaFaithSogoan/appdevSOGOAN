```python
Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

Enter Hours: 45 Enter Rate: 10 Pay: 475.0
```

```python

hours = float(input('Enter Hours: ')) # Get the number of hours worked from the user

# Get the hourly rate from the user
rate = float(input('Enter Rate: '))

# Calculate the gross pay
if hours <= 40:
    # If the number of hours worked is less than or equal to 40, pay the regular hourly rate
    pay = hours * rate
else:
    # If the number of hours worked is greater than 40, pay 1.5 times the hourly rate for the hours worked above 40
    pay = 40 * rate + (hours - 40) * rate * 1.5

print('Pay:', pay) # Print the gross pay
```

**CHALLENGE**

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0

```python
# Get the score from the user
score = float(input('Enter a score between 0 and 100: '))

# Determine the grade based on the score
if score >= 97 and score <= 100: #If the score is between 97 and 100 .
    grade = 1.0 #print 1.0 grade
elif score >= 93 and score < 97:  #If the score is between 93 and 97
    grade = 1.25  #print 1.25 grade
elif score >= 89 and score < 93: #If the score is between 89 and 93
    grade = 1.5 #print 1.5 grade
elif score >= 85 and score < 89: #If the score is between 85 and 89
    grade = 1.75  #print 1.75 grade
elif score >= 81 and score < 85: #If the score is between 81 and 85
    grade = 2.0  #print 2.0 grade
elif score >= 77 and score < 81: # If the score is between 77 and 81
    grade = 2.25  #print 2.25 grade
elif score >= 74 and score < 77: #If the score is between 74 and 77
    grade = 2.5  #print 2.5 grade
elif score >= 0 and score < 74: #If the score is between 0 and 74
    grade = 3.0   #print 3.0 grade
else:
    # If the score is not between 0 and 100, print an error message
    print('Invalid score')
    grade = None

# Print the grade
if grade is not None:
    print('Grade:', grade)
```
