Investigate on the difference between is and == given the following variables.

a = 2
b = 2.0
c = a

```python
 # Define two variables with the same value
a = 2
b = 2.0

# Test if a and b refer to the same object
print(a is b)  # Output: False

# Test if the values of a and b are equal
print(a == b)  # Output: True

# Define a new variable c that refers to the same object as a
c = a

# Test if c refers to the same object as a
print(c is a)  # Output: True

# Test if the values of c and b are equal
print(c == b)  # Output: True
```

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

Enter a number: 4
Enter a number: 5
Enter a number: bad data
Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333

```python
# Initialize the total and count variables to 0
total = 0
count = 0

# Prompt the user to enter a number
while True:
    # Get the user's input
    response = input('Enter a number: ')

    # Check if the user entered "done"
    if response.lower() == 'done':
        # If so, break out of the loop
        break

    # Try to convert the user's input to a float
    try:
        num = float(response)
    # If the conversion fails, print an error message and continue to the next iteration of the loop
    except ValueError:
        print('Invalid input')
        continue

    # Add the number to the total
    total += num

    # Increment the count
    count += 1

# Calculate the average
average = total / count

# Print the total, count, and average
print('Total:', total)
print('Count:', count)
print('Average:', average)
In this code, the user is repeatedly prom
```

Ch**allenge 2**

Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
# Initialize the minimum and maximum variables to the largest and smallest possible float values
minimum = float('inf')
maximum = float('-inf')

# Prompt the user to enter a number
while True:
    # Get the user's input
    response = input('Enter a number: ')

    # Check if the user entered "done"
    if response.lower() == 'done':
        # If so, break out of the loop
        break

    # Try to convert the user's input to a float
    try:
        num = float(response)
    # If the conversion fails, print an error message and continue to the next iteration of the loop
    except ValueError:
        print('Invalid input')
        continue

    # Update the minimum and maximum variables
    if num < minimum:
        minimum = num
    if num > maximum:
        maximum = num

# Print the minimum and maximum
print('Minimum:', minimum)
print('Maximum:', maximum)
```
