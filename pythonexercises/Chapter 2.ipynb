EXERCISE 2

**Write a program to prompt the user for hours and rate per hour to compute gross pay.**

Enter Hours: 35

Enter Rate: 2.75

Pay: 96.25

```python
hours = float(input("Enter Hours: ")) # Get the number of hours worked as a floating-point number
rate = float(input("Enter Rate: ")) # Get the hourly rate of pay as a floating-point number

# Calculate the amount of pay by multiplying the number of hours worked by the hourly rate
pay = hours * rate

print(f"Pay: {pay:.2f}") # Print the calculated pay with two decimal places using an f-string
```
